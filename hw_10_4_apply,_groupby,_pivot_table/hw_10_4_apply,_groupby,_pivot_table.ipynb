1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
df.groupby('Branch')['Total'].sum()
```

```python
df.groupby('Branch')['Total'].sum().nlargest(1)
```

2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  
Знайдіть відповідь за допомогою функціоналу Pandas.

```python
df.groupby(['City','Branch'])['Total'].sum().idxmax()
```

Філіал з найвищим прибутком знаходиться у місті Naypyitaw, це філіал С


```python
df.groupby('City').agg(avg_check=('Total','mean'),transactions=('Invoice ID','count'))
```

Також, в місті Naypyitaw найвищій середній чек, тому тут можна розмітисти ще один магазин

3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index = ['Branch'], columns= ['Product line'], values = 'Invoice ID', aggfunc= "count")
invoices_by_category
```

1. Яка категорія товарів продається найгірше?
   

```python
df.groupby('Product line')['Total'].sum().sort_values()
```

Найгірше продається Health and beauty Health and beauty, пропоную зробити акції та зменшити закупівлю товарів

Чи є категорія з високим середнім чеком, але низькою кількістю продажів?

```python
df.groupby('Product line')['Total'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```

Категорія Home and lifestyle з високим середнім чеком, але низькою кількістю продажів, це товари преміальні
Можна запіустити маркетингову підтримку та розмістити товар на видному місціб щоб збільшити продажі

3. Чи залежить середній чек від типу клієнта (Member vs Normal)?

```python
df.groupby('Customer type')['Total'].agg(['mean', 'sum', 'count']).sort_values('mean', ascending=False)
```

Member витрачають більше потрібно розвивати програму лояльності
